# 1 Distance from home by job role and attrition

```python
df1 = df[['DistanceFromHome','JobRole','Attrition']]
df1
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='JobRole', y='DistanceFromHome', hue='Attrition',data=df1)
plt.title("A graph showing the Distance from Home by Job Role and Attrition")
plt.xlabel("Job Role")
plt.ylabel("Distance from Home")
```

# 2) Average Monthly Income by education and by attrition

```python
MonthlyRevenue = df[['MonthlyIncome','Education','Attrition']]
MonthlyRevenue
```

```python
MonthlyRevenue = df.groupby(['Education','Attrition'])['MonthlyIncome'].mean().reset_index()
MonthlyRevenue
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Education', y='MonthlyIncome', hue='Attrition',data=MonthlyRevenue)
plt.title("A graph showing the Average Monthly Income by Education and Attrition")
plt.xlabel('Education')
plt.ylabel("Monthly Income")
```
